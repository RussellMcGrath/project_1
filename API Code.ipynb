```python
#import dependencies
import pandas as pd
import requests
import json
import time
import math
from pprint import pprint
from config import EDkey
```

```python
#create url building blocks
starter = "https://api.data.gov/ed/collegescorecard/v1/schools.json?"
results_per_page = "per_page=100"
api_key = f"api_key={EDkey}"
fields = ("_fields="
          "school.name,"
          "2018.student.size,"
          "school.online_only,"
          "2018.admissions.sat_scores.25th_percentile.critical_reading,"
          "2018.admissions.sat_scores.midpoint.critical_reading,"
          "2018.admissions.sat_scores.75th_percentile.critical_reading,"
          "2018.admissions.sat_scores.25th_percentile.math,"
          "2018.admissions.sat_scores.midpoint.math,"
          "2018.admissions.sat_scores.75th_percentile.math,"
          "2018.cost.tuition.in_state,"
          "2018.cost.tuition.out_of_state,"
          "2018.cost.tuition.program_year,"
          "2007.earnings.10_yrs_after_entry.working_not_enrolled.mean_earnings,"
          "2007.earnings.10_yrs_after_entry.median,"
          "2007.earnings.10_yrs_after_entry.working_not_enrolled.std_dev,"
          "2007.earnings.10_yrs_after_entry.mean_earnings.female_students,"
          "2007.earnings.10_yrs_after_entry.mean_earnings.male_students,"
          "2012.earnings.6_yrs_after_entry.working_not_enrolled.mean_earnings,"
          "2012.earnings.6_yrs_after_entry.median,"
          "2012.earnings.6_yrs_after_entry.working_not_enrolled.std_dev,"
          "2012.earnings.6_yrs_after_entry.mean_earnings.female_students,"
          "2012.earnings.6_yrs_after_entry.mean_earnings.male_students,"
          "location.lat,"
          "location.lon,"
          "school.locale,"
          "school.region_id,"
          "school.state,"
          "school.city,"
          "school.zip,"
          "school.institutional_characteristics.level,"
          "school.ownership"
         )
#construct url from variables
url = f"{starter}{fields}&{results_per_page}&{api_key}"

#run and print a test request for reference
response = requests.get(url).json()
pprint(response)
#record number of pages and records in response
total_pages = math.ceil((response["metadata"]["total"])/100)
total_records = response["metadata"]["total"]
```

```python
#since results come in one page at a time, create a for loop to request data from each page individually

#intialize list to house the data
big_results = []
#loop through each page of data
for page in range(total_pages):
    #define the string to be added to the url with the page requested
    page_code = f"page={page}"
    #construct url
    loop_url = f"{starter}{fields}&{results_per_page}&{page_code}&{api_key}"
    #record and json-ify the response
    loopresult = requests.get(loop_url).json()
    #add response to the running list
    big_results.extend(loopresult["results"])
    #display status of the loop
    print(f"Processing page {page} of {total_pages-1}")    
```

```python
#now we need to loop through the reponses and pull organize the data

#initialize list to house data for each field
schools = []
sizes = []
online_only = []
SAT_read_25 = []
SAT_read_50 = []
SAT_read_75 = []
SAT_math_25 = []
SAT_math_50 = []
SAT_math_75 = []
tuit_in = []
tuit_out = []
tuit_prog = []
earn_10_mean = []
earn_10_med = []
earn_10_std = []
earn_10_mean_female = []
earn_10_mean_male = []
earn_6_mean = []
earn_6_med = []
earn_6_std = []
earn_6_mean_female = []
earn_6_mean_male = []
lats = []
lons = []
locales = []
regions = []
states = []
zips = []
cities = []
inst_level = []
ownership = []

#loop through all the response records and organize the data into lists
for x in range(total_records):
    sizes.append(big_results[x]["2018.student.size"])
    schools.append(big_results[x]["school.name"])
    online_only.append(big_results[x]["school.online_only"])
    SAT_read_25.append(big_results[x]["2018.admissions.sat_scores.25th_percentile.critical_reading"])
    SAT_read_50.append(big_results[x]["2018.admissions.sat_scores.midpoint.critical_reading"])
    SAT_read_75.append(big_results[x]["2018.admissions.sat_scores.75th_percentile.critical_reading"])
    SAT_math_25.append(big_results[x]["2018.admissions.sat_scores.25th_percentile.math"])
    SAT_math_50.append(big_results[x]["2018.admissions.sat_scores.midpoint.math"])
    SAT_math_75.append(big_results[x]["2018.admissions.sat_scores.75th_percentile.math"])
    tuit_in.append(big_results[x]["2018.cost.tuition.in_state"])
    tuit_out.append(big_results[x]["2018.cost.tuition.out_of_state"])
    tuit_prog.append(big_results[x]["2018.cost.tuition.program_year"])
    earn_10_mean.append(big_results[x]["2007.earnings.10_yrs_after_entry.working_not_enrolled.mean_earnings"])
    earn_10_med.append(big_results[x]["2007.earnings.10_yrs_after_entry.median"])
    earn_10_std.append(big_results[x]["2007.earnings.10_yrs_after_entry.working_not_enrolled.std_dev"])
    earn_10_mean_female.append(big_results[x]["2007.earnings.10_yrs_after_entry.mean_earnings.female_students"])
    earn_10_mean_male.append(big_results[x]["2007.earnings.10_yrs_after_entry.mean_earnings.male_students"])
    earn_6_mean.append(big_results[x]["2012.earnings.6_yrs_after_entry.working_not_enrolled.mean_earnings"])
    earn_6_med.append(big_results[x]["2012.earnings.6_yrs_after_entry.median"])
    earn_6_std.append(big_results[x]["2012.earnings.6_yrs_after_entry.working_not_enrolled.std_dev"])
    earn_6_mean_female.append(big_results[x]["2012.earnings.6_yrs_after_entry.mean_earnings.female_students"])
    earn_6_mean_male.append(big_results[x]["2012.earnings.6_yrs_after_entry.mean_earnings.male_students"])
    lats.append(big_results[x]["location.lat"])
    lons.append(big_results[x]["location.lon"])
    locales.append(big_results[x]["school.locale"])
    regions.append(big_results[x]["school.region_id"])
    states.append(big_results[x]["school.state"])
    zips.append(big_results[x]["school.zip"])
    cities.append(big_results[x]["school.city"])
    inst_level.append(big_results[x]["school.institutional_characteristics.level"])
    ownership.append(big_results[x]["school.ownership"])

#create dataframe with data relevant to SAT analysis    
SAT_analysis = pd.DataFrame({"School":schools,
                             "Online only?":online_only,
                             "School size (students)":sizes,
                             "2yr/4yr":inst_level,
                             "sat_scores.25th_percentile.critical_reading":SAT_read_25,
                             "sat_scores.midpoint.critical_reading":SAT_read_50,
                             "sat_scores.75th_percentile.critical_reading":SAT_read_75,
                             "sat_scores.25th_percentile.math":SAT_math_25,
                             "sat_scores.midpoint.math":SAT_math_50,
                             "sat_scores.75th_percentile.math":SAT_math_75,
                             "Tuition (in state)":tuit_in,
                             "Tuition (out of state)":tuit_out,
                             "Tuition (program year)":tuit_prog,
                             "Pubic or Private":ownership
                            })

#create dataframe with data relevant to Earnings analysis
earnings_analysis = pd.DataFrame({"School":schools,
                                  "State":states,
                                  "Online only?":online_only,
                                  "School size (students)":sizes,
                                  "2yr/4yr":inst_level,
                                  "Pubic or Private":ownership,
                                  "Tuition (in state)":tuit_in,
                                  "Tuition (out of state)":tuit_out,
                                 "Earnings after 10 years - mean":earn_10_mean,
                                 "Earnings after 10 years - median":earn_10_med,
                                 "Earnings after 10 years - std":earn_10_std,
                                 "Earnings after 10 years - mean - female":earn_10_mean_female,
                                 "Earnings after 10 years - mean - male":earn_10_mean_male,
                                 "Earnings after 6 years - mean":earn_6_mean,
                                 "Earnings after 6 years - median":earn_6_med,
                                 "Earnings after 6 years - std":earn_6_std,
                                 "Earnings after 6 years - mean - female":earn_6_mean_female,
                                 "Earnings after 6 years - mean - male":earn_6_mean_male
                                 })

#create dataframe with data relevant to location analysis
school_locations = pd.DataFrame({"School":schools,
                                "City":cities,
                                "State":states,
                                "Zip":zips,
                                "Lat":lats,
                                "Lon":lons,
                                "Region":regions,
                                "Locale":locales,
                                 "2yr/4yr":inst_level,
                                 "Online only?":online_only,
                                  "School size (students)":sizes,
                                 "Pubic or Private":ownership
                                 })
```

```python
#export dataframes to csv files
SAT_analysis.to_csv("Docs/SAT Data.csv",index=False)
earnings_analysis.to_csv("Docs/Earnings Data.csv",index=False)
school_locations.to_csv("Docs/Location Data.csv",index=False)
```
