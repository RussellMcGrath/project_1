```python
starter = "https://api.data.gov/ed/collegescorecard/v1/schools.json?"
fields = ("_fields="
          "school.name,"
          "2018.student.size,"
          "2018.admissions.sat_scores.25th_percentile.critical_reading,"
          "2018.admissions.sat_scores.midpoint.critical_reading,"
          "2018.admissions.sat_scores.75th_percentile.critical_reading,"
          "2018.admissions.sat_scores.25th_percentile.math,"
          "2018.admissions.sat_scores.midpoint.math,"
          "2018.admissions.sat_scores.75th_percentile.math,"
          )

results_per_page = "per_page=100"
api_key = f"api_key={EDkey}"

url = f"{starter}{fields}&{results_per_page}&{api_key}"

#base_url = f"https://api.data.gov/ed/collegescorecard/v1/schools.json?_fields=school.name,2018.student.size&per_page=100&api_key={EDkey}"
response = requests.get(url).json()
pprint(response)

total_pages = math.ceil((response["metadata"]["total"])/100)
total_records = response["metadata"]["total"]
#pprint(response)
```

```python
big_results = []
for page in range(total_pages):
    page_code = f"page={page}"
    loop_url = f"{starter}{fields}&{results_per_page}&{page_code}&{api_key}"
    loopresult = requests.get(loop_url).json()
    big_results.extend(loopresult["results"])
    print(page)    
```

```python

schools = []
sizes = []
SAT_read_25 = []
SAT_read_50 = []
SAT_read_75 = []
SAT_math_25 = []
SAT_math_50 = []
SAT_math_75 = []
SAT_write_25 = []
SAT_write_50 = []
SAT_write_75 = []
#pprint(big_results)
for x in range(total_records):
    sizes.append(big_results[x]["2018.student.size"])
    schools.append(big_results[x]["school.name"])
    SAT_read_25.append(big_results[x]["2018.admissions.sat_scores.25th_percentile.critical_reading"])
    SAT_read_50.append(big_results[x]["2018.admissions.sat_scores.midpoint.critical_reading"])
    SAT_read_75.append(big_results[x]["2018.admissions.sat_scores.75th_percentile.critical_reading"])
    SAT_math_25.append(big_results[x]["2018.admissions.sat_scores.25th_percentile.math"])
    SAT_math_50.append(big_results[x]["2018.admissions.sat_scores.midpoint.math"])
    SAT_math_75.append(big_results[x]["2018.admissions.sat_scores.75th_percentile.math"])

    
SAT_df = pd.DataFrame({"School":schools,
                   "sat_scores.25th_percentile.critical_reading":SAT_read_25,
                   "sat_scores.midpoint.critical_reading":SAT_read_50,
                   "sat_scores.75th_percentile.critical_reading":SAT_read_75,
                   "sat_scores.25th_percentile.math":SAT_math_25,
                   "sat_scores.midpoint.math":SAT_math_50,
                   "sat_scores.75th_percentile.math":SAT_math_75,
                  })
SAT_df
```
