```python
#import the dependencies
import pandas as pd
from pandas import DataFrame
import numpy as np
import seaborn as sns
import scipy.stats as st
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create the dataframe
university_df = pd.read_csv('./Docs/Earnings Data.csv')
university_df.dropna(inplace = True)
university_df.drop_duplicates()
university_df
```

```python
#organized the dataframe in terms of higest school size to lowest
university_df['School size (students)']=university_df['School size (students)'].astype(float)
university_df['Earnings after 10 years - mean']=university_df['Earnings after 10 years - mean'].astype(float)
earnings = university_df.sort_values(['School size (students)','Earnings after 10 years - mean'], ascending=[False, False])
```

```python
earnings
```

```python
#wanted to look at the five universities with the higest school sized
earnings.head(5)
```

```python
#wanted to look at the five universities with the lowest school sized
earnings.tail(5)
```

```python
#create a scatter plot between the vaiables of school size and the mean earnings after 10 years
f = plt.figure(figsize=(20,10))
ax = f.add_subplot(111, facecolor=".1")
plt.scatter(university_df['School size (students)'],university_df['Earnings after 10 years - mean'], 
            c=university_df['Earnings after 10 years - mean'], cmap='Spectral', alpha=1, s=20)
correlation = st.pearsonr(university_df['School size (students)'],university_df['Earnings after 10 years - mean'])
plt.colorbar()
plt.title('SchoolSize_MeanEarnings')
plt.xlabel('X - School Size')
plt.ylabel('Y - Earnings after 10 years - mean')
print(f"The correlation between both factors is {round(correlation[0],2)}")
plt.savefig("Figures/Fig 1 - SchoolSize_MeanEarningAfter10Years")
plt.show()
```

```python
#limit the axis to see how the plot looks without possible outliers  
f = plt.figure(figsize=(20,10))
ax = f.add_subplot(111, facecolor=".1")
plt.scatter(university_df['School size (students)'],university_df['Earnings after 10 years - mean'], 
            c=university_df['Earnings after 10 years - mean'], cmap='Spectral', alpha=1, s=20)
correlation = st.pearsonr(university_df['School size (students)'],university_df['Earnings after 10 years - mean'])
plt.colorbar()
plt.title('SchoolSize_MeanEarnings')
plt.xlabel('X - School Size')
plt.ylabel('Y - Earnings after 10 years - mean')
print(f"The correlation between both factors is {round(correlation[0],2)}")
plt.xlim(-1000,30000)
plt.ylim(1000,100000)
plt.savefig("Figures/Fig 2 - SchoolSize_MeanEarningAfter10Years")
plt.show()
```
